Drop null and unwanted columns that were not helping in our analysis

```python
unwanted_cols = [
    'Product Description', 'Product Status', 'Customer Password', 'Customer Email', 'Category Id', 'Customer Id', 'Department Id', 'Order Customer Id', 
    'Order Item Cardprod Id', 'Order Item Id', 'Product Card Id', 'Product Category Id', 'Customer Fname', 'Customer Lname', 'Customer Street',
    'Product Image', 'Order Zipcode'
]
df=df.drop(columns=unwanted_cols)
```

```python
df.info()
```

Renaming column names 

```python
df.columns=df.columns.str.lower().str.replace(' ','_').str.replace('(','').str.replace(')','')
                      
```

```python
df.head()
```

```python
df.order_date_dateorders 
```

convert order_date column to proper format so that we can extract different parts from that column like month , year etc

```python
df['order_date_dateorders'] = pd.to_datetime(df['order_date_dateorders'])
```

```python
df.order_date_dateorders 
```

```python
df['order_year'] = df['order_date_dateorders'].dt.year
df['order_month'] = df['order_date_dateorders'].dt.month_name()
```

```python
df=df.drop(columns=['benefit_per_order'])
```

Round the floating values to 2 decimal places

```python
float_cols=df.select_dtypes(include=['float64']).columns
df[float_cols] = df[float_cols].round(2)
```

calculate actual delay in shipping

```python
df['actual_delay']=df['days_for_shipping_real']-df['days_for_shipment_scheduled']
```

```python
df.head()
```

Fixing encoding issues

```python
def fix_char(text):
    if not isinstance(text, str):
        return text
    try:
        return text.encode('latin1').decode('utf-8')
    except (UnicodeEncodeError, UnicodeDecodeError):
        return text

string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    df[col] = df[col].apply(fix_char)

print(df['order_state'].unique()[:5])
```

Loading data to postgres sql database for further insights

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv()
user=os.getenv('DB_USER')
password=os.getenv('DB_PASSWORD')
host=os.getenv('DB_HOST')
port=os.getenv('DB_PORT')
db_name=os.getenv('DB_NAME')

engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{db_name}')
```

```python
table_name='shipping'

df.to_sql(table_name, engine, if_exists='replace', index=False, chunksize=10000, method='multi')

print(f"Data successfully loaded into '{table_name}' in database '{db_name}'.")
```
